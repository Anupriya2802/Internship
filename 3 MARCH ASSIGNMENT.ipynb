```python
from bs4 import BeautifulSoup
import requests

url = ("https://www.cnbc.com/world/?region=world")
r = requests.get(url)
data = BeautifulSoup(r.text, 'html.parser')
```

```python
heading = []
for i in data.find_all("div",class_='RiverHeadline-headline RiverHeadline-hasThumbnail'):
    heading.append(i.text)
    heading
```

```python
heading[:]
```

```python
date = []
for i in data.find_all("time"):
    date.append(i.text)
    date
```

```python
from bs4 import BeautifulSoup
import requests

url = (" https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/ ")
r = requests.get(url)
data = BeautifulSoup(r.text, 'html.parser')
```

```python
article=data.find_all("div", {"class":"panelled-container mb-xl-0"})
```

```python
title = [data.find("h2", {"class":"h5 article-title"}).text for i in article]
date = [data.find("p", {"class":"article-date"}).text for i in article]
author = [data.find("p",{"class":"article-authors"}).text for i in article]
```
